```python
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}

# Part 2 - Task1
print("="*50)
print("Task 1 — Explore the Menu")
print("="*50)

# Part 2 - Task1.1
categories = set()
for item in menu.values():
    categories.add(item["category"])

for category in categories:
    print(f"===== {category} =====")

    for name, details in menu.items():
        if details["category"] == category:

            price = details["price"]
            status = "Available" if details["available"] else "Unavailable"

            print(f"{name:<15} ₹{price:<7.2f} [{status}]")

# Part 2 - Task1.2

# Total items on menu
print("="*50)
total_items = len(menu)
print("Total items on menu:", total_items)

# Total available items
available_count = 0
for item in menu.values():
    if item["available"]:
        available_count += 1

print("Total available items:", available_count)

# Most expensive item
max_price = 0
max_item = ""

for name, details in menu.items():
    if details["price"] > max_price:
        max_price = details["price"]
        max_item = name

print("Most expensive item:", max_item, f"(₹{max_price})")

# Items under ₹150
print("Items under ₹150:")
for name, details in menu.items():
    if details["price"] < 150:
        print(f"{name} (₹{details['price']})")


################################################################################

# Part 2 - Task2
print("="*50)
print("Task 2 — Cart Operations")
print("="*50)

cart = []

def show_cart():
    print("Current Cart:")
    if not cart:
        print("Cart is empty")
    else:
        for item in cart:
            print(f"{item['item']} x{item['quantity']} (₹{item['price']})")

# Part 2 - Task 2.1
def add_to_cart(item_name, quantity):

    # Check if item exists
    if item_name not in menu:
        print(f"{item_name} does not exist in menu.")
        return

    # Check availability
    if not menu[item_name]["available"]:
        print(f"{item_name} is currently unavailable.")
        return

    # Check if already in cart
    for item in cart:
        if item["item"] == item_name:
            item["quantity"] += quantity
            print(f"Updated {item_name} quantity to {item['quantity']}")
            return

    # Add new item
    cart.append({
        "item": item_name,
        "quantity": quantity,
        "price": menu[item_name]["price"]
    })

    print(f"Added {item_name} x{quantity} to cart")

# Part 2 - Task 2.2

# Remove item
def remove_from_cart(item_name):
    for item in cart:
        if item["item"] == item_name:
            cart.remove(item)
            print(f"{item_name} removed from cart")
            return

    print(f"{item_name} not found in cart")

# Part 2 - Task 2.3

# Update quantity
def update_quantity(item_name, quantity):
    for item in cart:
        if item["item"] == item_name:
            item["quantity"] = quantity
            print(f"{item_name} quantity updated to {quantity}")
            return

    print(f"{item_name} not found in cart")


# Part 2 - Task 2.4

# 1
add_to_cart("Paneer Tikka", 2)
show_cart()

# 2
add_to_cart("Gulab Jamun", 1)
show_cart()

# 3
add_to_cart("Paneer Tikka", 1)
show_cart()

# 4
add_to_cart("Mystery Burger", 1)
show_cart()

# 5
add_to_cart("Chicken Wings", 1)
show_cart()

# 6
remove_from_cart("Gulab Jamun")
show_cart()

# Part 2 - Task 2.5

# Final Order Summary
print("="*50)
print("Order Summary")
print("="*50)

subtotal = 0

for item in cart:
    item_total = item["quantity"] * item["price"]
    subtotal += item_total

    print(f"{item['item']:<18} x{item['quantity']}   ₹{item_total:.2f}")

print("-"*50)

gst = subtotal * 0.05
total = subtotal + gst

print(f"{'Subtotal:':<25} ₹{subtotal:.2f}")
print(f"{'GST (5%):':<25} ₹{gst:.2f}")
print(f"{'Total Payable:':<25} ₹{total:.2f}")
print("="*50)

################################################################################

# Part 2 - Task3

print("="*50)
print("Task 3 — Inventory Tracker with Deep Copy")
print("="*50)

import copy

# Part 2 - Task3.1

inventory_backup = copy.deepcopy(inventory)

# Change one value to test
inventory["Paneer Tikka"]["stock"] = 999

print("After modifying inventory:")
print("Inventory:", inventory["Paneer Tikka"])
print("Backup   :", inventory_backup["Paneer Tikka"])

# Restore original
inventory = copy.deepcopy(inventory_backup)


# Part 2 - Task3.2

print("Updating inventory based on cart...")

for item in cart:
    name = item["item"]
    qty_needed = item["quantity"]

    stock_available = inventory[name]["stock"]

    if stock_available >= qty_needed:
        inventory[name]["stock"] -= qty_needed
    else:
        print(f"Not enough stock for {name}. Only {stock_available} available.")
        inventory[name]["stock"] = 0


# Part 2 - Task3.3

print("Reorder Alerts:")
for name, details in inventory.items():
    stock = details["stock"]
    reorder_level = details["reorder_level"]

    if stock <= reorder_level:
        print(f"Reorder Alert: {name} — Only {stock} unit(s) left (reorder level: {reorder_level})")


# Part 2 - Task3.4

print("\nFinal Inventory (updated):")
for name, details in inventory.items():
    print(name, ":", details)

print("\nBackup Inventory (unchanged):")
for name, details in inventory_backup.items():
    print(name, ":", details)


################################################################################

# Part 2 - Task4

print("="*50)
print("Task 4 — Daily Sales Log Analysis")
print("="*50)

# Part 2 - Task4.1

print("Revenue Per Day:")
daily_revenue = {}

for date, orders in sales_log.items():
    total = 0
    for order in orders:
        total += order["total"]

    daily_revenue[date] = total
    print(f"{date} : ₹{total:.2f}")


# Part 2 - Task4.2

best_day = max(daily_revenue, key=daily_revenue.get)
print("Best-selling day:", best_day, f"(₹{daily_revenue[best_day]:.2f})")


# Part 2 - Task4.3

item_count = {}

for orders in sales_log.values():
    for order in orders:
        for item in order["items"]:
            item_count[item] = item_count.get(item, 0) + 1

most_ordered = max(item_count, key=item_count.get)
print("Most ordered item:", most_ordered, f"({item_count[most_ordered]} times)")


# Part 2 - Task4.4

sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

print("Updated Revenue Per Day:")

daily_revenue = {}

for date, orders in sales_log.items():
    total = sum(order["total"] for order in orders)
    daily_revenue[date] = total
    print(f"{date} : ₹{total:.2f}")

best_day = max(daily_revenue, key=daily_revenue.get)
print("Updated Best-selling day:", best_day, f"(₹{daily_revenue[best_day]:.2f})")


# Part 2 - Task4.5

print("All Orders:")
order_list = []

# Collect all orders
for date, orders in sales_log.items():
    for order in orders:
        order_list.append((date, order))

# Print using enumerate
for i, (date, order) in enumerate(order_list, start=1):
    items_str = ", ".join(order["items"])
    print(f"{i}. [{date}] Order #{order['order_id']} — ₹{order['total']:.2f} — Items: {items_str}")

```
